# 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

## 내가 풀어본 것

```python
prompt = """
1. 연락처 추가
2. 연락처 전체 보기
3. 검색: 이름을 입력 받아 전화번호 조회
4. 수정: 이름을 입력 받아 전화번호 수정
5. 삭제: 이름을 입력 받아 삭제
6. 종료

사용할 기능을 입력해주세요: """


choice=0
tel_book = {}

while choice != 6:
    print(prompt)
    choice = int(input())
    
    if choice == 1:
        info = input("이름과 연락처를 입력해주세요. (예: 홍길동 010-123-4567): ")
        name, num = info.split()
        tel_book[name] = num
        print("연락처가 추가되었습니다.")

    elif choice == 2:
        print("<<연락처>>")
        print(tel_book)
    
    elif choice == 3:
        name = input("검색할 이름을 입력해주세요: ")
        print(f"{name}의 전화번호는 {tel_book[name]} 입니다.")
    
    elif choice == 4:
        info = input("연락처를 수정할 사람의 이름과 전화번호을 입력해주세요: ")
        name, num = info.split()
        bf_num = tel_book[name]
        tel_book[name] = num
        print(f"{name}의 전화번호가 {bf_num} 에서 {num}으로 변경되었습니다.")
    
    elif choice == 5:
        name = input("연락처를 삭제할 사람의 이름을 입력해주세요: ")
        del tel_book[name]
        print("연락처가 삭제되었습니다.")
        
    else:
        print("프로그램을 종료합니다.")
        
```

## 같이 풀어보기

### 요구사항 분석 및 설계
요구사항 분석 -> 구체적인 동작을 정의    
I/O 설계

* 홍길동, 010-123-4567 -> 딕셔너리 {key:value}
* Input을 받아야겠구만 -> str 
* 프로그램 -> While
* 기능 1, 2, 3, 4, 5 -> if문. 숫자를 입력했을 때, 해당 기능 동작
----------------
* 기능 1 연락처 추가: 딕셔너리 추가. dict[key]=value
* 기능 2 전체보기: for문 돌려서 출력
* 기능 3 조회: 이름 입력했을 때 전화번호 출력.
* 기능 4 수정: dict[key]=value
* 기능 5 삭제: del
* 기능 6 종료: break
* 예외처리



### 조립하기

```python
contact = {} # 전화번호부 만들기

while True:
    print('----- 전화번호부 프로그램 실행 -----')
    print('1.추가 2.보기 3.검색 4.수정 5.삭제 9.종료')
    menu = int(input('메뉴 선택: '))
    
    try: 
        if menu == 1:
            print("연락처 추가 작업 실행")
            new_name = input("이름: ")
            new_tel = input("전화번호: ")
            contact.setdefault(new_name, new_tel)


        elif menu == 2:
            print("연락처 보기 작업 실행")
            for name, tel in contact.items():
                print(name, ":", tel)

        elif menu == 3:
            print("연락처 검색 작업 실행")
            search_name = input("검색 이름: ")
            print(contact.get(search_name, '없는 연락처입니다.'))
            # 인덱스로 조회하지 않는 이유? 해당 인덱스가 없으면 오류가 남. --> get 함수 사용

        elif menu == 4:
            print("연락처 수정 작업 실행")
            mod_name = input("수정할 이름:")
            if mod_name in contact:
                mod_tel = input("새 전화번호: ")
                contact[mod_name] = mod_tel
            else:
                print("등록되지 않은 이름입니다. ")

        elif menu == 5:
            print("연락처 삭제 작업 실행")
            del_name = input("삭제할 이름: ")

            # If 연락처 저장소에 삭제 이름이 존재하면, 저장소에서 삭제
            if del_name in contact: 
                del contact[del_name]
            else:
                print("등록되지 않은 이름입니다.")


        elif menu == 9:
            print("연락처 종료 작업 실행")
            break

        else:
            print("잘못된 메뉴 선택입니다.")
    except:
        print("예외가 발생했습니다.")
```
